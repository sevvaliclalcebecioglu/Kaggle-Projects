## *NLP â€“ Text Cleaning (Overview)*

```python
df["clean_overview"] = df["Overview"].astype(str)
```

```python
df["clean_overview"] = df["clean_overview"].apply(nfx.remove_stopwords)
```

```python
df["clean_overview"] = df["clean_overview"].apply(nfx.remove_special_characters)
```

```python
df["clean_overview"] = df["clean_overview"].apply(nfx.remove_punctuations)
```

## *Word Count Distribution*

```python
df["word_count"] = df["clean_overview"].apply(lambda x: len(x.split()))

plt.figure()
sns.histplot(df["word_count"])
plt.title("Word Count Distribution")
plt.show()
```
